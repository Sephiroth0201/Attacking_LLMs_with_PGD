```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
gpt_model = AutoModelForCausalLM.from_pretrained("gpt2", pad_token_id=50256)
gpt_tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
gpt_prompt = "Reporter: 'Thank you for joining us today. Can you tell us about the moment you realized you had superpowers?' Hero: 'It all started when I was a kid and one day I"


gpt_inputs = gpt_tokenizer(gpt_prompt, return_tensors="pt")

gpt_input_ids=gpt_inputs.input_ids
attention_mask=gpt_inputs.attention_mask

gpt_gen_tokens = gpt_model.generate(
    gpt_input_ids,
    attention_mask=attention_mask,
    do_sample=True,
    temperature=0.9,
    max_length=200,
    pad_token_id=50256
)
gpt_gen_text = gpt_tokenizer.batch_decode(gpt_gen_tokens,skip_special_tokens=True)[0]
```

```python
print(gpt_gen_text)
```
